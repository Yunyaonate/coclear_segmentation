```python
import cv2
import numpy as np
import os
from PIL import Image

def preprocess_mask(img_path, save_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    # Find the maximum pixel value in the image
    max_val = np.max(img)

    # Normalize the pixel values to the range [0, 1]
    max_val = np.max(img)
    normalised_img = (img / max_val * 255).astype(np.uint8)

    # Create a copy of the normalized image for filtering
    fil_img = normalised_img.copy()

    # Threshold the filtered image
    lighter = fil_img >= 150
    darker = fil_img < 150

    # Set lighter pixels to white and darker pixels to black
    fil_img[lighter] = 255
    fil_img[darker] = 0
    

    # Construct the save path and filename
    filename = os.path.basename(img_path)
    save_filename = f"{os.path.splitext(filename)[0]}_mask.png"
    save_file_path = os.path.join(save_path, save_filename)

    # Save the filtered image
    cv2.imwrite(save_file_path, fil_img)

    return save_file_path  # Return the path of the saved image

# Usage
# original_images_folder = '/Users/yunyao/Desktop/Capstone/Data/Test_205/edited_filtered'
# filtered_images_folder = '/Users/yunyao/Desktop/Capstone/Data/Test_205/mask'
original_images_folder = '/Users/yunyao/Desktop/Capstone/Data/Cato_A Copy/edited_filtered'
filtered_images_folder = '/Users/yunyao/Desktop/Capstone/Data/Cato_A Copy/mask'
# Make sure the save directory exists
if not os.path.exists(filtered_images_folder):
    os.makedirs(filtered_images_folder)

# Process all images in the input directory
for filename in os.listdir(original_images_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        img_path = os.path.join(original_images_folder, filename)
        filtered_image_path = preprocess_mask(img_path, filtered_images_folder)
        if filtered_image_path:
            print(f"Processed and saved: {filtered_image_path}")
```

# Clean up the dir
After checking all the filtered images are edited and converted to binary image (only 0 and 255 for easy to visualise), we clean up the work space by:
- delete the temperate files
- delete the working space for manually editing the label 
- save the original files to dir img_dir and mask images to ann_dir for later use



```python
import os
import shutil

def clean_data(base_dir):
    """
    Cleans the data by deleting files that start with '._' or contain '_._' in the filename,
    and removes the 'filtered' or 'edited_filtered' directories.
    :param base_dir: Path to the base directory containing all data.
    """
    # Walk through the directory
    for dirpath, dirnames, filenames in os.walk(base_dir, topdown=False):  # topdown=False to allow deleting subdirectories
        for filename in filenames:
            # Check if the file starts with '._' or contains '_._'
            if filename.startswith('._') or '_._' in filename:
                file_to_delete = os.path.join(dirpath, filename)
                os.remove(file_to_delete)  # Delete the file
                print(f"Deleted file: {file_to_delete}")
        
        # Since we're modifying the list we're iterating over, we iterate over a copy of the list
        for dirname in dirnames[:]:
            # Check if the directory name is 'filtered' or 'edited_filtered'
            if dirname == 'filtered' or dirname == "edited_filtered":
                dir_to_delete = os.path.join(dirpath, dirname)
                shutil.rmtree(dir_to_delete)  # Delete the directory
                print(f"Deleted directory: {dir_to_delete}")

# Specify the base directory (where the CI Data directory is located)
base_dir = '/Users/yunyao/Desktop/Capstone/Data/Cato_A Copy'

# Run the function to clean the data
clean_data(base_dir)
```
